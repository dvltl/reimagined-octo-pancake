```python
class MyRegressor:
    pa = PassiveAggressiveRegressor()
    rr = Ridge()
    def fit(self, X, y):
        self.pa = self.pa.fit(X,y)
        self.rr = self.rr.fit(X,y)
    def predict(self, X):
        pa_labels = self.pa.predict(X)
        rr_labels = self.rr.predict(X)
        return np.array((pa_labels + rr_labels) / 2)
```
